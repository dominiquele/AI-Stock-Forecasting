```python
#Import the libraries

import yfinance as yf
#API Package that downloads daily stock & index prices
```

```python
#Initial ticker class to download price history for a single symbol
sp500 = yf.Ticker ("^GSPC")
#GSPC is the sp500 index
```

```python
#Query the historical prices of the index from the time it was created
sp500 =sp500.history(period="max")
```

```python
#Calling SP500 history (highest and lowest price)
sp500
```

```python
#Calling sp500 Index (Trading Date) (Help slice the data)
sp500.index
```

```python
#Plot the data on the dataset
sp500.plot.line(y="Close", use_index=True)
#Show the close date on the y-axis and trading date on the x-axis
```

```python
#Data Cleaning
del sp500["Dividends"]
del sp500["Stock Splits"]
#Remove the dividens and splits column (those columns are used for individual stocks)
```

```python
#Predict if the price will go up or down tomorrow
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
#Run the prediction
sp500
```

```python
#Set a target, which is the price the machine learning is trying to predict
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
#Is the target price for tomorrow greater than the one of today (The boolean is converted into an integer for the ML to use)
```

```python
#Run the dataset with the target column (1 = the price went up, 0 = the price went down)
sp500
```

```python
#Clean the data by removing price stocks that were before 1990
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
#Run the dataset with the price stock that start on 1990-01-01
sp500
```

```python
#Training the machine learning

from sklearn.ensemble import RandomForestClassifier
#Import RandomForestClassifier as the default model (Good model to pick up non linear graph)

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
#100 individual trees (smaller sample, but it allows to have quick results) and min of 100 to prevent overfit

train = sp500.iloc[:-100]
#learn how to predict the stock price
test = sp500.iloc[-100:]
#put all the rows except the last 100 rows

predictors = ["Close", "Volume", "Open", "High", "Low"]
#Create a list with the predictions
model.fit(train[predictors], train["Target"])
#Train the model with the predictors
```

```python
#Measure how accurate the model is
from sklearn.metrics import precision_score
#Import sklearn.metrics to verify the accuracy of the model

preds = model.predict(test[predictors])
#Generate the predictions
```

```python
# Run the predictions
preds
```

```python
#Turn the predictions into a panda series (easier to work with)
import pandas as pd
#Import pandas
preds = pd.Series(preds, index=test.index)
#Turn the predictions into panda series
```

```python
preds
```

```python
#Calculate the precision score
precision_score(test["Target"], preds)
```

```python
#The prediction score is only 55% accurate
```

```python
#Plot the prediction

combined = pd.concat([test["Target"], preds], axis=1)
#Combine the predicted and targeted value
```

```python
#Plot the combined data
combined.plot()
```

```python
# orange = the prediction
# blue = the results (what actually happened in the stock market)
```

```python
#Backtesting (Testing against other data set to improve the accuracy of the algorithm)

def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
  #Create a prediction function to consolidate all the algorithm into one function
```

```python
#Create the backstesting function

def backtest(data, model, predictors, start=2500, step=250):
  all_predictions = []
#The first model will take 2500days of data and train it with those 10 years of data
#Next, the model will take the value of the 10th first years (2500 days) and l will predict the value for the 11th years (the next 250 days or next year)
#Then, the model will take the value of the 11th first years and predict the values for ther 12th years and so on

#Create a function to loop across the data year by year
  for i in range(start, data.shape[0], step):
    train = data.iloc[0:i].copy()
    #Split the train & test set
    #The train set includes all the data before the current year (year we start to predict)
    test = data.iloc[i:(i+step)].copy()
    #The test set is the data set for this year (that we are trying to predict)
    predictions= predict(train, test, predictors, model)
    #Generate the predictions
    all_predictions.append(predictions)
    #Append/Add all the prediction for the given year to the data set
  return pd.concat(all_predictions)
  #Concatenate the predictions (take a list of data frames and combine them all into a single data frame)
```

```python
#Backtest the sp500
predictions = backtest(sp500, model, predictors)
```

```python
#Run how many time the model predicted that the stock would go up or down to evaluate the error of the predictions
predictions["Predictions"].value_counts()
```

```python
#Check the precision score to see how accurate the predictions were
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#Accross 6000 the model was accurate about 53% of the time
```

```python
#To figure out if the prediction is good or bad (benchmark/baseline) look at the % of days the stock actually went up
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
#The sp500 stock went up 53.4% and went down 46.5%
#The algorithm performed a little bit worst with 52.9% compared to the natural occurence percentage
```

```python
#Adding additional predictors to the model to improve the accuracy

horizons = [2,5,60,250,1000]
#Make the model compared the predictions (the stock going up) to 2 days ago, 5 days ago, etc.
#Giving more data to analyze for the model improve its accuracy
new_predictors =[]
#Add new columns with the new data set (predictions)

for horizon in horizons:
  rolling_averages = sp500.rolling(horizon).mean()
  #Create variety of rolling averages from the horizons (periods of data we are comparing too)
  ratio_column = f"Close_Ratio_{horizon}"
  sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
  #Create a column for each horizon (period) and compared the sp500 closing price to the horizon
  #The ratio demonstraste today's closing price compared to the closing price in the period given (2 days ago, 5 days ago, etc.)
  #Ratio includes the current day it calculates
  trend_column = f"Trend_{horizon}"
  sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
  #The trend column shows the number of days in the past x days (horizons) that the stock price went up
  #And it will find the rolling sum for the past x days (horizons)
  #The rolling sum is the number of days the price actually went up
  #Trend does not include the current day it calculates

  new_predictors += [ratio_column, trend_column]
  #Add those new data / predictors to the data set
```

```python
#Remove the extra NaN columns
sp500 = sp500.dropna()
```

```python
#The data now start in 1993 due to the trend column
#We needed 4 years of data to compute the trend calculation
```

```python
#Update the model
#Increase the number of estimators to 200 and decrease min_samples_split to 50

model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
#Improve the prediction function
# Provide more control to the predictions (going up = 1, going down = 0) with probablity
#The model will return the probablity that the stock will go down or up tomorrow

def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors]) [:,1]
  preds [preds >= .6] = 1
  preds [preds < .6] = 0
  #The default treshold is .5 so if there is a 50% change the price will go up the model will display it as a 1
  #We opted to a .6 to make the model more confident in its predictions
  #This will decrease the number of days the price will go up but it wll increase the chance that the price will go up
  #This model is made to recommend investors to buy the stock if there is a high chance that the price will go up, which is what this adjustment will do
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
  #Create a prediction function to consolidate all the algorithm into one function
```

```python
#Run the backtest again
predictions = backtest(sp500, model, new_predictors)
```

```python
#Count the number of days the stock goes up and the number of days the stock goes down
predictions["Predictions"].value_counts()
```

```python
#Now the number of days being predicted that the price will go up is lower 830 compared to before, which was 2596 days due to the treshold.
#This makes the predictions more accurate due to the model being more confident about the number of days where the stock goes up
```

```python
#Check the precision score to see if the improve model is more accurate than the old one
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#The precision score has increase from 52.8 to 57.2 percent
#This means that the model is more accurate
#This is a very good result if we take into consideration that we are just looking at the time series data and historical prices of the index
```
